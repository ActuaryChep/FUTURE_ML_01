```python
import pandas as pd #for loading and manipulating time series data.
import numpy as np #for numerical operations
from statsmodels.tsa.arima.model import ARIMA #the model class from statsmodels for time series forecasting.
import matplotlib.pyplot as plt #for plotting the data and forecast
from sklearn.metrics import mean_squared_error # to measure how accurate the forecast is compared to actual data.
```

```python
df = pd.read_csv("C:/Users/User/Downloads/mock_kaggle.csv") #loading data
```

```python
df['data'] = pd.to_datetime(df['data']) #converts data column (dates) into Python datetime objects.
```

```python
df.set_index('data', inplace=True) #Sets the data column as the DataFrame index so we can treat it as a time series.
```

```python
monthly_sales = df['venda'].resample('M').sum() #groups data by month.
```

```python
train = monthly_sales[:-6] #uses all data except the last 6 months for training the model.
test = monthly_sales[-6:] #uses the last 6 months for testing model accuracy.
```

```python
model = ARIMA(train, order=(21, 2, 0))#p=1: autoregressive lag of 1,d=1: first difference to make the series stationary,q=1: moving average lag of 1
```

```python
model_fit = model.fit() #Fits (trains) the ARIMA model on the train data.


```

```python
forecast = model_fit.forecast(steps=6)#Forecasts the next 6 months using the trained ARIMA model
```

```python
rmse = np.sqrt(mean_squared_error(test, forecast))
print(f"RMSE: {rmse:.2f}") # Evaluate Forecast Accuracy
```

```python
#Creates a visual plot of:Historical training data,Actual sales from the test period Forecasted sales
plt.figure(figsize=(10, 5))
plt.plot(train.index, train, label='Train')
plt.plot(test.index, test, label='Actual', color='green')
plt.plot(test.index, forecast, label='Forecast', color='red', linestyle='--')
plt.title('ARIMA Forecast vs Actual Sales')
plt.xlabel('Date')
plt.ylabel('Monthly Sales')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
